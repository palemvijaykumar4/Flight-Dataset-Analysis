```python
# with bar graphs
plt.figure(figsize = (16,4))
plt.subplot(1,2,1)
plt.bar(d['departure_time'].value_counts().index , d['departure_time'].value_counts().values,color='red')
plt.title('departure_Time')
plt.ylabel('D.freq')
plt.xlabel('D.Time')

plt.subplot(1,2,2)
plt.bar(d['arrival_time'].value_counts().index , d['arrival_time'].value_counts().values,color='yellow')
plt.title('arrival_Time')
plt.ylabel('A.freq')
plt.xlabel('A.Time')
plt.show()
```
